```python
import os
```

```python

def search(filename, text):
    temp = []
    with open(filename) as f:
        for line in f:
            if text in line:
                temp.append(float(line.strip(text)))
    return temp
```

```python
curr_dir = os.getcwd()
actual_dir = curr_dir + '/res18_more_pics/bceloss/logs/11166764.out'

loss_D_fake = search(actual_dir, 'loss_D_fake:')
loss_D_real = search(actual_dir, 'loss_D_real:')
loss_D = search(actual_dir, 'loss_D:')

loss_G_GAN = search(actual_dir, 'loss_G_GAN:')
loss_G_L1 = search(actual_dir, 'loss_G_L1:')
loss_G = search(actual_dir, 'loss_G:')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5,5), constrained_layout=True)
axs = fig.subplots(2)

fig.suptitle("Resnet 18 - 10k pictures - BCEloss")
axs[0].plot([i for i in range(0,len(loss_D_fake))], loss_D_fake, label='loss_D_fake')
axs[0].plot([i for i in range(0,len(loss_D_real))], loss_D_real, label='loss_D_real')
axs[0].set_title("Discriminator loss")
# axs[1].plot([i for i in range(0,len(loss_G_GAN))], loss_G_GAN, label='loss_G_GAN')
axs[1].plot([i for i in range(0,len(loss_G))], loss_G, label='loss_G')
axs[1].set_title("Generator loss")
axs[0].legend()
axs[0].grid()
axs[1].legend()
axs[1].grid()

plt.show()

fig.savefig(f"Graphs/res18_more_pics-bceloss.png")
```
